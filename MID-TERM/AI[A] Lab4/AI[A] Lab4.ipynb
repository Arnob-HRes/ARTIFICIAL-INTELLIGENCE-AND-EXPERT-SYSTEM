# Matrix (List of List)

```python
m=[
    [2,4],
    [10,20],
    [50,100]
]
print(type(m))
print(m)
print(m[0])
print(len(m))
print(len(m[0]))
print(m[0][1])
```

```python
for i in range(len(m)):
    for j in range(len(m[0])):
        print(m[i][j],end=" ")
    print()
```

```python
m=[]
row=3
col=2
for i in range(row):
    r=[]
    for j in range(col):
        r.append(0)
    m.append(r)

print(m)
        
```

```python
def createMatrix(row,col):
    m=[]
    for i in range(row):
        r=[]
        for j in range(col):
            r.append(0)
        m.append(r)
    return m

def printMatrix(m):
    for i in m:
        print(i)
        
M=createMatrix(10,20)
printMatrix(M)
```

```python
def createIdentityMatrix(n):
    m=[]
    for i in range(n):
        r=[]
        for j in range(n):
            if i==j:
                r.append(1)
            else:
                r.append(0)
        m.append(r)
    return m

g=createIdentityMatrix(10)
printMatrix(g)
```

```python
import random
def createMatrix(row,col):
    m=[]
    for i in range(row):
        r=[]
        for j in range(col):
            r.append(random.randint(1,1000))
        m.append(r)
    return m

def printMatrix(m):
    for i in m:
        print(i)
        
M=createMatrix(5,5)
printMatrix(M)
```
